```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataDir='/Users/Torben/Code/coronaweather/';

trainPath=dataDir+'training_data_with_weather_info_week_1.csv';
df=pd.read_csv(trainPath)

uniqueProvinces=list(set(df['Province/State']))
```

```python
# plot a province's precepitation against confirmed coronavirus cases 

# find province from list
myProvinceStr='Washington'
thisProvinceInd=uniqueProvinces.index(myProvinceStr);


# get the data
thisProvince=uniqueProvinces[thisProvinceInd]

thisProvinceInds=df.index[df['Province/State'] == thisProvince].tolist()

thisProvinceDates=df['Date'][thisProvinceInds]
thisProvinceConfirmed=df['ConfirmedCases'][thisProvinceInds]
# thisProvinceFatalities=df['Fatalities'][thisProvinceInds]
thisProvincePrecip=df['prcp'][thisProvinceInds]

# plot it
fig = plt.figure(figsize=(20,5));
ax1 = fig.add_subplot(111);

ax1.plot(thisProvinceDates, thisProvincePrecip, 'b-');
ax1.set_xlabel('precepitation')
ax1.set_ylabel('Date')
ax1.set_title(uniqueProvinces[thisProvinceInd])
plt.xticks(rotation=45);

ax2 = ax1.twinx()
ax2.plot(thisProvinceDates, thisProvinceConfirmed, 'r-');
ax2.set_ylabel('Confirmed Cases');
```
